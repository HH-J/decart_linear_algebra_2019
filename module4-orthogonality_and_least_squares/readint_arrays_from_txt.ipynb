```python
import csv
```

```python
def str2float(txt):
    try:
        return float(txt)
    except:
        return np.nan
```

```python
def read_data(fname, header=True):
    data = []
    with open(fname) as csvfile:
        reader = csv.reader(csvfile, delimiter=',', quotechar='|')
        if header:
            h = next(reader)
        else:
            h = []
        for row in reader:
            data.append(list(map(str2float,row)))
    return np.array(data), h
read_data('Class_Ozone_Data.csv')
```
